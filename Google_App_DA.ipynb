```python
# 9 dropping the rows containing 'Varies with device' in Size column
gdata=gdata[~(gdata['Size']=='Varies with device')]
```

```python
gdata
```

```python
#  Converting size column into suitable data type
def convert_size_to_int(size_str):
    if size_str.endswith('M'):
        return int(float(size_str.replace('M', '')) * 1e6)
    if size_str.endswith('K'):
        return int(float(size_str.replace('K', '')) * 1e3)

gdata['Size'] = gdata['Size'].apply(convert_size_to_int)

```

```python
gdata
```

```python
gdata['Size'].dtypes
```

```python
# 10 Replacing unwanted characters and treating data type
gdata['Installs']=gdata['Installs'].str.replace('[+,]','',regex=True)
```

```python
gdata['Installs']=pd.to_numeric(gdata['Installs'])
```

```python
# 11 Replacing unwanted characters and treating the Price column
gdata['Price']=gdata['Price'].str.replace('[$]','',regex=True)
```

```python
gdata['Price']=pd.to_numeric(gdata['Price'])
```
